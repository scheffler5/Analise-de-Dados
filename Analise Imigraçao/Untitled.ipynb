```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
arquivo = pd.read_csv("imigrantes_canada.csv")
arquivo
```

```python
arquivo.info()
```

```python
arquivo.set_index("País", inplace=True)
arquivo
```

```python
anos = list(map(str,range(1980,2014)))
```

```python
anos
```

```python
brasil = arquivo.loc["Brasil",anos]
```

```python
brasil_dic = {"Anos" : brasil.index.tolist(), "Imigrantes" : brasil.values.tolist()}
dados_brasil = pd.DataFrame(brasil_dic)
```

```python
dados_brasil
```

```python
america_sul = arquivo.query('Região == "América do Sul"').sort_values("Total", ascending=True)
america_sul
```

```python
plt.figure(figsize=(8,4))
plt.plot(dados_brasil["Anos"], dados_brasil["Imigrantes"])
plt.title("Imigraçao do Brasil para o Canada")
plt.xlabel("Ano")
plt.ylabel("Imigrantes")
plt.xticks(["1980","1985","1990","1995","2000", "2005","2010","2013"])
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(dados_brasil["Anos"],dados_brasil["Imigrantes"],lw=3, color="g")
ax.set_title("Imigraçao do Brasil para o Canada",fontsize=18,loc="left")

ax.set_xlabel("Ano", fontsize=14)
ax.set_ylabel("Numero de Imigrantes",fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.grid()
plt.show()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(15,5))
axs[0].plot(dados_brasil["Anos"],dados_brasil["Imigrantes"])
axs[0].set_title("Imigraçao do Brasil para o Canada")
axs[0].set_xlabel("Ano")
axs[0].set_ylabel("Imigrantes")
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].grid()


axs[1].boxplot(dados_brasil["Imigrantes"])
axs[1].set_title("Imigraçao do Brasil para o Canada")
axs[1].set_xlabel("Brasil")
axs[1].set_ylabel("Imigrantes")
axs[1].grid()
plt.show()
```

```python
fig, axs = plt.subplots(2,2, figsize=(10,6))
fig.subplots_adjust(hspace=0.5,wspace=0.3)
fig.suptitle("Imigraçao do quatro maiores paises para o canada")

axs[0,0].plot(arquivo.loc["Brasil",anos])
axs[0,0].set_title("Brasil")

axs[0,1].plot(arquivo.loc["Colômbia",anos])
axs[0,1].set_title("Colombia")

axs[1,0].plot(arquivo.loc["Argentina", anos])
axs[1,0].set_title("Argentina")

axs[1,1].plot(arquivo.loc["Peru",anos])
axs[1,1].set_title("Peru")

for ax in axs.flat:
    ax.xaxis.set_major_locator(plt.MultipleLocator(5))
for ax in axs.flat:
    ax.set_xlabel("Ano")
    ax.set_ylabel("Numero de imigrantes")

ymin = 0
ymax = 7000
for ax in axs.ravel():
    ax.set_ylim(ymin,ymax)
plt.show()
```

```python
cores = []
for pais in america_sul.index:
    if pais == "Brasil":
        cores.append("green")
    else:
        cores.append("silver")
fig, ax = plt.subplots(figsize=(12, 5))
ax.barh(america_sul.index, america_sul['Total'], color=cores)
ax.set_title("Imigraçao da america do sul para o Canada",loc="left",fontsize=14)
ax.set_ylabel("")
ax.set_xlabel("Numero de Imigrantes", fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

for i,v in enumerate(america_sul["Total"]):
    ax.text(v +20, i , str(v), color="Black", fontsize=10,ha="left", va= "center")

ax.set_frame_on(False)
ax.get_xaxis().set_visible(False)
ax.tick_params(axis="both", which="both", length=0)
fig.savefig("Imigraçao grafico de Barra.png", transparent= False, dpi=300,bbox_inches="tight")
plt.show()
```
