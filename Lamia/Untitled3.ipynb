```python
contagem = df.groupby(["Smoking_History", "Previous_Cancer_Diagnosis"]).size().unstack()
contagem.plot(kind="bar", figsize=(8, 6), stacked=False)
plt.title("Distribuição de Fumantes e Diagnosticos Previos", fontsize=16)
plt.xlabel("Fumantes", fontsize=14)
plt.ylabel("Contagem", fontsize=14)

plt.savefig("c:/Users/gabri/Downloads/Distribuição_de_Fumantes_e_Diagnosticos_previos.png")
plt.show()
```

```python
filtro = df.query('Smoking_History == "Current" and Previous_Cancer_Diagnosis == True and Years_Smoked <= 5')

# Agrupar os dados filtrados
contagem = filtro.groupby(["Smoking_History", "Previous_Cancer_Diagnosis"]).size().unstack()

# Criar o gráfico de barras
contagem.plot(kind="bar", figsize=(8, 6), stacked=False)

# Adicionar título e rótulos
plt.title("Distribuição de Fumantes (Current) e Diagnósticos Prévios (True)", fontsize=16)
plt.xlabel("Fumantes", fontsize=14)
plt.ylabel("Contagem", fontsize=14)

# Salvar o gráfico
plt.savefig("c:/Users/gabri/Downloads/Distribuição_de_Fumantes_e_Diagnosticos_previos_filtradoss.png", dpi=300)
plt.show()
```
