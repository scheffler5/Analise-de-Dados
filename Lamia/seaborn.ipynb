```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
import numpy as np
```

```python
x = np.linspace(1980, 2025, 10)  
mu1, sigma1 = 2000, 10
mu2, sigma2 = 2000, 5
linha1 = norm.pdf(x, mu1, sigma1) * 100
linha2 = norm.pdf(x, mu2, sigma2) * 100
data = pd.DataFrame({
    'Anos': np.tile(x, 2),  
    'Valores': np.concatenate([linha1, linha2]),  
    'Grupo': ['Década de 80'] * len(x) + ['Década dos anos 2000'] * len(x)  
})
plt.figure(figsize=(10, 6))
sns.barplot(data=data, x='Anos', y='Valores', hue='Grupo', palette='coolwarm')
plt.grid(axis='y')
plt.title('Gráfico de Consumidores de Programas de Televisão no Tempo')
plt.xlabel('Anos')
plt.ylabel('Idades')
plt.legend(title='Grupos')
plt.savefig('C:\\Users\\gabri\\Downloads\\Seaborn.png', format='png')

plt.show()
```
